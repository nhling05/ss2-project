```python
# Huấn Luyện Mô Hình Nhận Dạng Tiền Giấy Với YOLOv8
# Ngày tạo: 04-05-2025

import os
import yaml
import shutil
from IPython.display import clear_output, display, Image
from google.colab import drive, files
import numpy as np
import cv2
import zipfile
import matplotlib.pyplot as plt

# Hàm tiện ích để hiển thị trạng thái
def show_status(message):
    print(f"[INFO] {message}")

# Bước 1: Cài đặt các gói cần thiết trước khi import
show_status("Đang cài đặt các gói cần thiết...")
!pip install ultralytics
!pip install torch torchvision
clear_output(wait=True)
show_status("Đã cài đặt các gói cần thiết!")

# Bây giờ import các thư viện đã cài đặt
import torch
from ultralytics import YOLO

# Bước 2: Kết nối Google Drive để lưu trữ mô hình
show_status("Đang kết nối Google Drive...")
drive.mount('/content/drive')
clear_output(wait=True)
show_status("Google Drive đã được kết nối thành công!")

# Bước 3: Tạo thư mục cho dự án
project_dir = '/content/dollar_bill_detection'
show_status(f"Tạo thư mục dự án tại {project_dir}")
!mkdir -p {project_dir}
!mkdir -p {project_dir}/dataset
!mkdir -p {project_dir}/dataset/train/images
!mkdir -p {project_dir}/dataset/train/labels
!mkdir -p {project_dir}/dataset/valid/images
!mkdir -p {project_dir}/dataset/valid/labels
!mkdir -p {project_dir}/dataset/test/images
!mkdir -p {project_dir}/dataset/test/labels

# Bước 4: Upload dataset từ máy tính
show_status("Vui lòng upload các file dữ liệu hoặc tệp ZIP chứa dataset...")
print("Có 2 cách để upload dữ liệu:")
print("Cách 1: Upload file data.yaml")
print("Cách 2: Upload file ZIP chứa toàn bộ dataset")
print("\nVui lòng chọn một trong hai cách trên và thực hiện theo hướng dẫn.")

# Tạo một ô để người dùng upload file
uploaded = files.upload()

# Kiểm tra loại file được upload
uploaded_files = list(uploaded.keys())
yaml_file = None
zip_file = None

for filename in uploaded_files:
    if filename.endswith('.yaml'):
        yaml_file = filename
        # Di chuyển file YAML vào thư mục dataset
        shutil.move(filename, f"{project_dir}/dataset/data.yaml")
        show_status(f"Đã tìm thấy file YAML: {filename}")
    elif filename.endswith('.zip'):
        zip_file = filename
        show_status(f"Đã tìm thấy file ZIP: {filename}")

# Xử lý file ZIP nếu có
if zip_file:
    show_status(f"Đang giải nén {zip_file}...")
    with zipfile.ZipFile(zip_file, 'r') as zip_ref:
        zip_ref.extractall(f"{project_dir}/dataset")
    show_status("Đã giải nén xong!")

    # Kiểm tra cấu trúc thư mục sau khi giải nén
    if not os.path.exists(f"{project_dir}/dataset/data.yaml"):
        show_status("Không tìm thấy file data.yaml trong thư mục dataset. Vui lòng upload file này.")
        yaml_file = files.upload()
        for filename in yaml_file.keys():
            if filename.endswith('.yaml'):
                shutil.move(filename, f"{project_dir}/dataset/data.yaml")
                show_status(f"Đã di chuyển file {filename} vào thư mục dataset")

# Nếu không có file ZIP, yêu cầu upload thêm các thư mục dữ liệu
if not zip_file:
    show_status("Vui lòng upload thư mục train/images")
    train_images = files.upload()
    for filename in train_images.keys():
        shutil.move(filename, f"{project_dir}/dataset/train/images/{filename}")

    show_status("Vui lòng upload thư mục train/labels")
    train_labels = files.upload()
    for filename in train_labels.keys():
        shutil.move(filename, f"{project_dir}/dataset/train/labels/{filename}")

    show_status("Vui lòng upload thư mục valid/images")
    valid_images = files.upload()
    for filename in valid_images.keys():
        shutil.move(filename, f"{project_dir}/dataset/valid/images/{filename}")

    show_status("Vui lòng upload thư mục valid/labels")
    valid_labels = files.upload()
    for filename in valid_labels.keys():
        shutil.move(filename, f"{project_dir}/dataset/valid/labels/{filename}")

# Bước 5: Kiểm tra cấu trúc dataset
show_status("Kiểm tra cấu trúc dataset...")
!ls -la {project_dir}/dataset

if os.path.exists(f"{project_dir}/dataset/data.yaml"):
    show_status("Đọc thông tin từ file data.yaml")
    with open(f"{project_dir}/dataset/data.yaml", 'r') as file:
        yaml_content = yaml.safe_load(file)
        print("\nThông tin classes:")
        for idx, name in enumerate(yaml_content.get('names', [])):
            print(f"  {idx}: {name}")

        # Cập nhật đường dẫn trong file YAML nếu cần
        if 'train' in yaml_content and not yaml_content['train'].startswith('/'):
            yaml_content['train'] = f"{project_dir}/dataset/train/images"
        if 'val' in yaml_content and not yaml_content['val'].startswith('/'):
            yaml_content['val'] = f"{project_dir}/dataset/valid/images"
        if 'test' in yaml_content and not yaml_content['test'].startswith('/'):
            yaml_content['test'] = f"{project_dir}/dataset/test/images"

        # Lưu lại file YAML đã cập nhật
        with open(f"{project_dir}/dataset/data.yaml", 'w') as outfile:
            yaml.dump(yaml_content, outfile, default_flow_style=False)

        show_status("Đã cập nhật đường dẫn trong file data.yaml")
else:
    show_status("CẢNH BÁO: Không tìm thấy file data.yaml!")
    print("Vui lòng upload file data.yaml để tiếp tục.")
    exit()

# Bước 6: Thiết lập các tham số huấn luyện
show_status("Thiết lập các tham số huấn luyện...")

# Cho phép người dùng tùy chỉnh các tham số
print("\nBạn có thể tùy chỉnh các tham số huấn luyện sau:")
print("1. Số lượng epochs (mặc định: 100)")
print("2. Kích thước ảnh (mặc định: 640)")
print("3. Kích thước batch (mặc định: 16)")
print("4. Model size (mặc định: n - nano)")

epochs = 100
img_size = 640
batch_size = 16
model_size = 'n'  # n (nano), s (small), m (medium), l (large), x (xlarge)

# Người dùng có thể tùy chỉnh trong đoạn code này
# epochs = int(input("Nhập số epochs: ") or "100")
# img_size = int(input("Nhập kích thước ảnh: ") or "640")
# batch_size = int(input("Nhập batch size: ") or "16")
# model_size = input("Nhập model size (n/s/m/l/x): ") or "n"

model_path = f'yolov8{model_size}.pt'
show_status(f"Sử dụng mô hình: {model_path}")
show_status(f"Số lượng epochs: {epochs}")
show_status(f"Kích thước ảnh: {img_size}")
show_status(f"Batch size: {batch_size}")

# Bước 7: Tải mô hình YOLOv8 pretrained
show_status("Đang tải mô hình YOLOv8 pretrained...")
model = YOLO(model_path)
show_status("Đã tải mô hình thành công!")

# Bước 8: Bắt đầu huấn luyện
show_status("Bắt đầu quá trình huấn luyện...")
show_status("Quá trình này có thể mất từ vài phút đến vài giờ tùy thuộc vào kích thước dataset và GPU được phân bổ.")
show_status("Bạn có thể theo dõi tiến trình huấn luyện ở đây.")

# Kiểm tra xem GPU có sẵn không
device = 0 if torch.cuda.is_available() else 'cpu'
show_status(f"Đang sử dụng thiết bị: {'GPU' if device == 0 else 'CPU'}")

results = model.train(
    data=f"{project_dir}/dataset/data.yaml",
    epochs=epochs,
    imgsz=img_size,
    batch=batch_size,
    name='dollar_bill_model',
    plots=True,
    device=device
)

show_status("Huấn luyện hoàn tất!")

# Bước 9: Đánh giá mô hình
show_status("Đánh giá mô hình trên dữ liệu validation...")
model.val()

# Bước 10: Vẽ biểu đồ kết quả huấn luyện
show_status("Hiển thị kết quả huấn luyện...")
try:
    results_path = '/content/runs/detect/dollar_bill_model'
    img = Image(filename=f'{results_path}/results.png', width=800)
    display(img)
    show_status("Biểu đồ kết quả huấn luyện đã được hiển thị ở trên.")
except Exception as e:
    show_status(f"Không thể hiển thị biểu đồ: {e}")

# Bước 11: Xuất mô hình sang các định dạng khác nhau
show_status("Đang xuất mô hình sang các định dạng khác nhau...")

# Xuất sang định dạng ONNX
show_status("Xuất sang định dạng ONNX...")
onnx_success = model.export(format='onnx')
if onnx_success:
    show_status("Xuất sang ONNX thành công!")
else:
    show_status("Không thể xuất sang định dạng ONNX!")

# Xuất sang định dạng TorchScript
show_status("Xuất sang định dạng TorchScript...")
torchscript_success = model.export(format='torchscript')
if torchscript_success:
    show_status("Xuất sang TorchScript thành công!")
else:
    show_status("Không thể xuất sang định dạng TorchScript!")

# Bước 12: Kiểm tra mô hình trên ảnh thử nghiệm
show_status("Kiểm tra mô hình trên ảnh thử nghiệm...")
test_images_dir = f"{project_dir}/dataset/test/images"
if os.path.exists(test_images_dir) and len(os.listdir(test_images_dir)) > 0:
    test_images = os.listdir(test_images_dir)[:5]  # Lấy tối đa 5 ảnh để kiểm tra

    plt.figure(figsize=(20, 15))
    for i, img_name in enumerate(test_images):
        img_path = os.path.join(test_images_dir, img_name)
        results = model.predict(img_path, conf=0.25)

        # Hiển thị ảnh với các dự đoán
        res_plotted = results[0].plot()

        plt.subplot(len(test_images), 1, i+1)
        plt.imshow(cv2.cvtColor(res_plotted, cv2.COLOR_BGR2RGB))
        plt.title(f'Ảnh thử nghiệm {i+1}: {img_name}')
        plt.axis('off')

    plt.tight_layout()
    plt.show()
    show_status("Đã hiển thị kết quả dự đoán trên ảnh thử nghiệm.")
else:
    show_status("Không tìm thấy thư mục ảnh thử nghiệm hoặc thư mục trống.")

# Bước 13: Lưu mô hình vào Google Drive
show_status("Đang lưu mô hình vào Google Drive...")
model_save_path = '/content/drive/MyDrive/dollar_bill_model'
!mkdir -p {model_save_path}
!cp -r /content/runs/detect/dollar_bill_model/* {model_save_path}/

# Tìm đúng đường dẫn đến file mô hình tốt nhất
best_model_path = '/content/runs/detect/dollar_bill_model/weights/best.pt'
if os.path.exists(best_model_path):
    !cp {best_model_path} {model_save_path}/best.pt
    show_status(f"Đã sao chép file mô hình tốt nhất vào: {model_save_path}/best.pt")
else:
    # Thử tìm các file .pt khác
    show_status("Đang tìm file mô hình .pt...")
    !find /content/runs/detect/dollar_bill_model -name "*.pt"

    # Lấy file cuối cùng (.pt) nếu có
    last_pt_files = !find /content/runs/detect/dollar_bill_model -name "*.pt" | tail -1
    if len(last_pt_files) > 0 and last_pt_files[0]:
        last_pt_file = last_pt_files[0]
        !cp {last_pt_file} {model_save_path}/best.pt
        show_status(f"Đã sao chép file mô hình từ {last_pt_file} vào: {model_save_path}/best.pt")
    else:
        show_status("CẢNH BÁO: Không tìm thấy file mô hình .pt nào!")

# Bước 14: Tải trọng số mô hình về máy tính
show_status("Đang chuẩn bị tải trọng số mô hình về máy tính...")
if os.path.exists(f"{model_save_path}/best.pt"):
    !cp {model_save_path}/best.pt ./best_dollar_bill_model.pt
    files.download("best_dollar_bill_model.pt")
    show_status("Đã tải file mô hình về máy tính thành công!")
else:
    show_status("CẢNH BÁO: Không tìm thấy file mô hình để tải về!")

show_status("Đã tải các trọng số mô hình về máy tính của bạn!")

# Bước 15: Hiển thị ma trận nhầm lẫn
show_status("Hiển thị ma trận nhầm lẫn...")
try:
    conf_matrix_path = f'{results_path}/confusion_matrix.png'
    if os.path.exists(conf_matrix_path):
        img = Image(filename=conf_matrix_path, width=800)
        display(img)
        show_status("Ma trận nhầm lẫn đã được hiển thị ở trên.")
    else:
        show_status("Không tìm thấy ma trận nhầm lẫn.")
except Exception as e:
    show_status(f"Không thể hiển thị ma trận nhầm lẫn: {e}")

# Kết thúc
show_status("Quá trình huấn luyện và đánh giá đã hoàn tất!")
show_status("Bạn có thể sử dụng mô hình đã tải về trong hệ thống Vision-I của mình.")
show_status("QUAN TRỌNG: Mô hình đã được lưu vào Google Drive và tải về máy tính của bạn.")
```
