```python
#-------------------------------------------------------------------------------
# B∆∞·ªõc 0: C√°c h√†m ti·ªán √≠ch v√† import ban ƒë·∫ßu
#-------------------------------------------------------------------------------
import os
import yaml
import shutil
from IPython.display import clear_output, display # Image s·∫Ω ƒë∆∞·ª£c import sau khi c√†i ultralytics
from google.colab import drive, files
import numpy as np
import cv2 # S·∫Ω ƒë∆∞·ª£c c√†i c√πng opencv-python t·ª´ ultralytics
import zipfile
import matplotlib.pyplot as plt # S·∫Ω ƒë∆∞·ª£c c√†i c√πng ultralytics

def show_status(message, is_warning=False, is_error=False):
    prefix = "[INFO]"
    if is_warning:
        prefix = "[C·∫¢NH B√ÅO]"
    elif is_error:
        prefix = "[L·ªñI]"
    print(f"{prefix} {message}")

#-------------------------------------------------------------------------------
# B∆∞·ªõc 1: C√†i ƒë·∫∑t c√°c g√≥i c·∫ßn thi·∫øt
#-------------------------------------------------------------------------------
show_status("ƒêang c√†i ƒë·∫∑t c√°c g√≥i c·∫ßn thi·∫øt (ultralytics, torch, torchvision)...")
%pip install ultralytics torch torchvision --quiet
from IPython.display import Image # Import Image sau khi ultralytics (v√† pillow) ƒë∆∞·ª£c c√†i
clear_output(wait=True) # X√≥a output c·ªßa pip install
show_status("ƒê√£ c√†i ƒë·∫∑t c√°c g√≥i c·∫ßn thi·∫øt!")

# Import c√°c th∆∞ vi·ªán ƒë√£ c√†i ƒë·∫∑t
import torch
from ultralytics import YOLO

#-------------------------------------------------------------------------------
# B∆∞·ªõc 2: K·∫øt n·ªëi Google Drive
#-------------------------------------------------------------------------------
show_status("ƒêang k·∫øt n·ªëi Google Drive...")
try:
    drive.mount('/content/drive', force_remount=True) # Th√™m force_remount ƒë·ªÉ ƒë·∫£m b·∫£o k·∫øt n·ªëi l·∫°i n·∫øu c·∫ßn
    clear_output(wait=True)
    show_status("Google Drive ƒë√£ ƒë∆∞·ª£c k·∫øt n·ªëi th√†nh c√¥ng!")
except Exception as e:
    show_status(f"Kh√¥ng th·ªÉ k·∫øt n·ªëi Google Drive: {e}", is_error=True)
    show_status("Vui l√≤ng ki·ªÉm tra l·∫°i quy·ªÅn truy c·∫≠p ho·∫∑c th·ª≠ k·∫øt n·ªëi l·∫°i.", is_warning=True)
    # C√≥ th·ªÉ d·ª´ng ·ªü ƒë√¢y ho·∫∑c cho ph√©p ti·∫øp t·ª•c m√† kh√¥ng c√≥ Drive
    # exit()

#-------------------------------------------------------------------------------
# B∆∞·ªõc 3: T·∫°o th∆∞ m·ª•c cho d·ª± √°n VND
#-------------------------------------------------------------------------------
# ƒê·ªïi t√™n th∆∞ m·ª•c d·ª± √°n ƒë·ªÉ ph√¢n bi·ªát v·ªõi d·ª± √°n USD
project_dir_vnd = '/content/vnd_currency_detection'
dataset_base_dir_vnd = os.path.join(project_dir_vnd, 'dataset') # Th∆∞ m·ª•c ch·ª©a data.yaml v√† c√°c folder train/valid/test

show_status(f"T·∫°o th∆∞ m·ª•c d·ª± √°n t·∫°i {project_dir_vnd}")
os.makedirs(dataset_base_dir_vnd, exist_ok=True) # T·∫°o c·∫£ th∆∞ m·ª•c dataset


#-------------------------------------------------------------------------------
# B∆∞·ªõc 4: Upload Dataset Ti·ªÅn VND
#-------------------------------------------------------------------------------
show_status("Vui l√≤ng upload dataset ti·ªÅn VND.")

uploaded = files.upload()

main_zip_file = None
data_yaml_file = None
uploaded_file_names = list(uploaded.keys())

for filename in uploaded_file_names:
    if filename.lower().endswith('.zip'): # Ki·ªÉm tra kh√¥ng ph√¢n bi·ªát hoa th∆∞·ªùng
        # Gi·∫£ ƒë·ªãnh file zip ch√≠nh ch·ª©a to√†n b·ªô dataset n·∫øu c√≥ t√™n chung chung
        # Ho·∫∑c ng∆∞·ªùi d√πng ch·ªâ upload 1 file zip
        if main_zip_file is None : # ∆Øu ti√™n file zip ƒë·∫ßu ti√™n l√†m file zip ch√≠nh
             main_zip_file = filename
        show_status(f"ƒê√£ t√¨m th·∫•y file ZIP: {filename}")
    elif filename.lower() == 'data.yaml':
        data_yaml_file = filename
        # Di chuy·ªÉn file YAML v√†o th∆∞ m·ª•c dataset_base_dir_vnd n·∫øu n√≥ ch∆∞a ·ªü ƒë√≥
        if os.path.dirname(os.path.join(os.getcwd(), filename)) != dataset_base_dir_vnd:
            shutil.move(filename, os.path.join(dataset_base_dir_vnd, 'data.yaml'))
        show_status(f"ƒê√£ t√¨m th·∫•y v√† di chuy·ªÉn file YAML: {filename} v√†o {dataset_base_dir_vnd}")

# X·ª≠ l√Ω file ZIP ch√≠nh n·∫øu c√≥
if main_zip_file:
    show_status(f"ƒêang gi·∫£i n√©n {main_zip_file} v√†o {dataset_base_dir_vnd}...")
    try:
        with zipfile.ZipFile(main_zip_file, 'r') as zip_ref:
            zip_ref.extractall(dataset_base_dir_vnd) # Gi·∫£i n√©n tr·ª±c ti·∫øp v√†o th∆∞ m·ª•c dataset
        show_status(f"ƒê√£ gi·∫£i n√©n {main_zip_file} th√†nh c√¥ng!")

        # Sau khi gi·∫£i n√©n, ki·ªÉm tra xem data.yaml c√≥ n·∫±m trong th∆∞ m·ª•c con kh√¥ng
        # V√≠ d·ª•: dataset.zip gi·∫£i n√©n ra dataset/data.yaml (thay v√¨ data.yaml tr·ª±c ti·∫øp)
        # T√¨m file data.yaml trong c√°c th∆∞ m·ª•c con v·ª´a gi·∫£i n√©n
        extracted_yaml_path = None
        for root, dirs, files_in_root in os.walk(dataset_base_dir_vnd):
            if 'data.yaml' in files_in_root:
                extracted_yaml_path = os.path.join(root, 'data.yaml')
                break

        if extracted_yaml_path and extracted_yaml_path != os.path.join(dataset_base_dir_vnd, 'data.yaml'):
            # Di chuy·ªÉn data.yaml ra th∆∞ m·ª•c dataset_base_dir_vnd n·∫øu n√≥ n·∫±m s√¢u h∆°n
            shutil.move(extracted_yaml_path, os.path.join(dataset_base_dir_vnd, 'data.yaml'))
            show_status(f"ƒê√£ di chuy·ªÉn data.yaml t·ª´ {extracted_yaml_path} ƒë·∫øn th∆∞ m·ª•c g·ªëc c·ªßa dataset.")
            # X√≥a th∆∞ m·ª•c r·ªóng c√≤n l·∫°i n·∫øu c√≥ (v√≠ d·ª•: n·∫øu data.yaml n·∫±m trong m·ªôt th∆∞ m·ª•c con duy nh·∫•t)
            if os.path.isdir(os.path.dirname(extracted_yaml_path)) and not os.listdir(os.path.dirname(extracted_yaml_path)):
                os.rmdir(os.path.dirname(extracted_yaml_path))

        data_yaml_path_check = os.path.join(dataset_base_dir_vnd, 'data.yaml')
        if not os.path.exists(data_yaml_path_check):
            show_status("Kh√¥ng t√¨m th·∫•y 'data.yaml' trong file ZIP ho·∫∑c ·ªü th∆∞ m·ª•c dataset g·ªëc sau khi gi·∫£i n√©n.", is_warning=True)
            show_status("Vui l√≤ng upload 'data.yaml' ri√™ng.")
            uploaded_yaml = files.upload()
            if 'data.yaml' in uploaded_yaml:
                shutil.move('data.yaml', data_yaml_path_check)
                data_yaml_file = 'data.yaml' # C·∫≠p nh·∫≠t l·∫°i t√™n file yaml
                show_status(f"ƒê√£ di chuy·ªÉn 'data.yaml' v√†o {dataset_base_dir_vnd}")
            else:
                show_status("Kh√¥ng upload file 'data.yaml'. Qu√° tr√¨nh c√≥ th·ªÉ g·∫∑p l·ªói.", is_error=True)
                # exit()
        else:
            data_yaml_file = 'data.yaml' # ƒê√°nh d·∫•u l√† ƒë√£ c√≥ data.yaml

    except zipfile.BadZipFile:
        show_status(f"File {main_zip_file} kh√¥ng ph·∫£i l√† file ZIP h·ª£p l·ªá ho·∫∑c b·ªã l·ªói.", is_error=True)
        main_zip_file = None # ƒê·∫∑t l·∫°i ƒë·ªÉ c√≥ th·ªÉ x·ª≠ l√Ω theo c√°ch upload file r·ªùi
    except Exception as e:
        show_status(f"L·ªói khi gi·∫£i n√©n {main_zip_file}: {e}", is_error=True)
        main_zip_file = None

# N·∫øu kh√¥ng c√≥ file ZIP ch√≠nh HO·∫∂C data.yaml v·∫´n ch∆∞a c√≥
if not data_yaml_file:
    data_yaml_path_check = os.path.join(dataset_base_dir_vnd, 'data.yaml')
    if not os.path.exists(data_yaml_path_check):
        show_status("Ch∆∞a c√≥ file 'data.yaml'. Vui l√≤ng upload 'data.yaml'.")
        uploaded_yaml = files.upload()
        if 'data.yaml' in uploaded_yaml:
            shutil.move('data.yaml', data_yaml_path_check)
            data_yaml_file = 'data.yaml'
            show_status(f"ƒê√£ di chuy·ªÉn 'data.yaml' v√†o {dataset_base_dir_vnd}")
        else:
            show_status("B·∫ÆT BU·ªòC ph·∫£i c√≥ 'data.yaml'. D·ª´ng ch∆∞∆°ng tr√¨nh.", is_error=True)
            exit()
    else:
         data_yaml_file = 'data.yaml' # ƒê√£ c√≥ s·∫µn t·ª´ tr∆∞·ªõc


#-------------------------------------------------------------------------------
# B∆∞·ªõc 5: Ki·ªÉm tra v√† C·∫≠p nh·∫≠t file data.yaml
#-------------------------------------------------------------------------------
path_to_data_yaml = os.path.join(dataset_base_dir_vnd, 'data.yaml')
show_status(f"Ki·ªÉm tra file data.yaml t·∫°i: {path_to_data_yaml}")

if os.path.exists(path_to_data_yaml):
    show_status("ƒê·ªçc th√¥ng tin t·ª´ file data.yaml...")
    try:
        with open(path_to_data_yaml, 'r') as file:
            yaml_content = yaml.safe_load(file)

        if yaml_content is None:
            show_status("File data.yaml r·ªóng ho·∫∑c kh√¥ng ƒë√∫ng ƒë·ªãnh d·∫°ng.", is_error=True)
            exit()

        print("\nTh√¥ng tin classes t·ª´ data.yaml:")
        names = yaml_content.get('names', [])
        if names:
            for idx, name in enumerate(names):
                print(f"  {idx}: {name}")
        else:
            show_status("Kh√¥ng t√¨m th·∫•y danh s√°ch 'names' trong data.yaml.", is_warning=True)

        nc = yaml_content.get('nc')
        if nc is not None:
            print(f"S·ªë l∆∞·ª£ng classes (nc): {nc}")
            if names and len(names) != nc:
                show_status(f"S·ªë l∆∞·ª£ng 'names' ({len(names)}) kh√¥ng kh·ªõp v·ªõi 'nc' ({nc}).", is_warning=True)
        else:
            show_status("Kh√¥ng t√¨m th·∫•y 'nc' (s·ªë l∆∞·ª£ng classes) trong data.yaml.", is_warning=True)


        requires_update = False
        for key in ['train', 'val', 'test']:
            if key in yaml_content:
                current_path = yaml_content[key]
                # N·∫øu ƒë∆∞·ªùng d·∫´n l√† t∆∞∆°ng ƒë·ªëi v√† kh√¥ng b·∫Øt ƒë·∫ßu b·∫±ng '../' (ch·ªâ t√™n th∆∞ m·ª•c)
                # th√¨ n√≥ ƒë∆∞·ª£c hi·ªÉu l√† n·∫±m trong c√πng th∆∞ m·ª•c v·ªõi data.yaml
                # V√≠ d·ª•: train: train/images

                # ƒê∆∞·ªùng d·∫´n tuy·ªát ƒë·ªëi ƒë·∫øn th∆∞ m·ª•c images/labels mong mu·ªën
                # V√≠ d·ª•: /content/vnd_currency_detection/dataset/train/images
                expected_abs_path_dir = os.path.join(dataset_base_dir_vnd, os.path.normpath(current_path))

                if not os.path.exists(expected_abs_path_dir):
                    show_status(f"ƒê∆∞·ªùng d·∫´n '{current_path}' cho '{key}' trong data.yaml kh√¥ng t·ªìn t·∫°i t·∫°i '{expected_abs_path_dir}'.", is_warning=True)
                    # C√≥ th·ªÉ th·ª≠ m·ªôt v√†i c·∫•u tr√∫c ph·ªï bi·∫øn
                    # V√≠ d·ª•, n·∫øu data.yaml n·∫±m trong dataset/ v√† dataset.yaml ghi train: images/ (thi·∫øu train/)
                    common_paths_to_try = [
                        os.path.join(dataset_base_dir_vnd, key, 'images'), # dataset/train/images
                        os.path.join(dataset_base_dir_vnd, os.path.basename(os.path.normpath(current_path))), # dataset/images (n·∫øu current_path l√† train/images)
                    ]
                    found_alternative = False
                    for alt_path_dir in common_paths_to_try:
                        if os.path.exists(alt_path_dir) and os.path.isdir(alt_path_dir):
                            # C·∫≠p nh·∫≠t l·∫°i ƒë∆∞·ªùng d·∫´n t∆∞∆°ng ƒë·ªëi cho data.yaml
                            # v√≠ d·ª•: t·ª´ dataset_base_dir_vnd, ƒë∆∞·ªùng d·∫´n ƒë·∫øn alt_path_dir l√† g√¨?
                            relative_new_path = os.path.relpath(alt_path_dir, start=dataset_base_dir_vnd)
                            yaml_content[key] = relative_new_path # S·ª≠a th√†nh ƒë∆∞·ªùng d·∫´n t∆∞∆°ng ƒë·ªëi ƒë√∫ng
                            show_status(f"ƒê√£ c·∫≠p nh·∫≠t ƒë∆∞·ªùng d·∫´n cho '{key}' th√†nh '{relative_new_path}' (tr·ªè ƒë·∫øn {alt_path_dir})")
                            requires_update = True
                            found_alternative = True
                            break
                    if not found_alternative:
                         show_status(f"Kh√¥ng th·ªÉ t·ª± ƒë·ªông s·ª≠a ƒë∆∞·ªùng d·∫´n cho '{key}'. Vui l√≤ng ki·ªÉm tra th·ªß c√¥ng file data.yaml.", is_error=True)

        if requires_update:
            with open(path_to_data_yaml, 'w') as outfile:
                yaml.dump(yaml_content, outfile, default_flow_style=False, sort_keys=False)
            show_status("File data.yaml ƒë√£ ƒë∆∞·ª£c c·∫≠p nh·∫≠t v·ªõi ƒë∆∞·ªùng d·∫´n m·ªõi (n·∫øu c√≥).")

        # In l·∫°i c·∫•u tr√∫c th∆∞ m·ª•c dataset ƒë·ªÉ ki·ªÉm tra
        show_status("C·∫•u tr√∫c th∆∞ m·ª•c dataset hi·ªán t·∫°i:")
        # !ls -R {dataset_base_dir_vnd} # ls -R c√≥ th·ªÉ qu√° d√†i, ch·ªâ hi·ªán c·∫•p ƒë·∫ßu
        for item in os.listdir(dataset_base_dir_vnd):
            print(f"  ‚îî‚îÄ‚îÄ {item}")
            if os.path.isdir(os.path.join(dataset_base_dir_vnd, item)):
                 for sub_item in os.listdir(os.path.join(dataset_base_dir_vnd, item)):
                      print(f"      ‚îî‚îÄ‚îÄ {sub_item}")


    except yaml.YAMLError as e:
        show_status(f"L·ªói khi ƒë·ªçc file data.yaml: {e}", is_error=True)
        exit()
    except Exception as e:
        show_status(f"L·ªói kh√¥ng x√°c ƒë·ªãnh khi x·ª≠ l√Ω data.yaml: {e}", is_error=True)
        exit()
else:
    show_status(f"KH√îNG T√åM TH·∫§Y file data.yaml t·∫°i '{path_to_data_yaml}'!", is_error=True)
    show_status("Vui l√≤ng upload file data.yaml v√†o th∆∞ m·ª•c dataset v√† ch·∫°y l·∫°i.", is_warning=True)
    exit()

#-------------------------------------------------------------------------------
# B∆∞·ªõc 6: Thi·∫øt l·∫≠p c√°c tham s·ªë hu·∫•n luy·ªán
#-------------------------------------------------------------------------------
show_status("Thi·∫øt l·∫≠p c√°c tham s·ªë hu·∫•n luy·ªán cho m√¥ h√¨nh VND...")

# Gi√° tr·ªã m·∫∑c ƒë·ªãnh
default_epochs = 75 # Gi·∫£m m·ªôt ch√∫t so v·ªõi 100 n·∫øu dataset nh·ªè ho·∫∑c ƒë·ªÉ test nhanh
default_img_size = 640
default_batch_size = 16 # Ph·ªï bi·∫øn cho Colab free tier
default_model_size = 'n' # nano (nh·∫π nh·∫•t), s (small), m (medium)

print("\nB·∫°n c√≥ th·ªÉ t√πy ch·ªânh c√°c tham s·ªë hu·∫•n luy·ªán (nh·∫•n Enter ƒë·ªÉ d√πng gi√° tr·ªã m·∫∑c ƒë·ªãnh):")
try:
    epochs_input = input(f"1. S·ªë l∆∞·ª£ng epochs (m·∫∑c ƒë·ªãnh: {default_epochs}): ")
    epochs = int(epochs_input) if epochs_input else default_epochs

    img_size_input = input(f"2. K√≠ch th∆∞·ªõc ·∫£nh (m·∫∑c ƒë·ªãnh: {default_img_size}): ")
    img_size = int(img_size_input) if img_size_input else default_img_size

    batch_size_input = input(f"3. K√≠ch th∆∞·ªõc batch (m·∫∑c ƒë·ªãnh: {default_batch_size}): ")
    batch_size = int(batch_size_input) if batch_size_input else default_batch_size

    model_size_input = input(f"4. Model size (n, s, m, l, x - m·∫∑c ƒë·ªãnh: {default_model_size}): ").lower()
    model_size = model_size_input if model_size_input in ['n', 's', 'm', 'l', 'x'] else default_model_size
except ValueError:
    show_status("ƒê·∫ßu v√†o kh√¥ng h·ª£p l·ªá, s·ª≠ d·ª•ng gi√° tr·ªã m·∫∑c ƒë·ªãnh.", is_warning=True)
    epochs, img_size, batch_size, model_size = default_epochs, default_img_size, default_batch_size, default_model_size


model_base_name = f'yolov8{model_size}.pt'
show_status(f"S·ª≠ d·ª•ng m√¥ h√¨nh c∆° s·ªü: {model_base_name}")
show_status(f"S·ªë l∆∞·ª£ng epochs: {epochs}")
show_status(f"K√≠ch th∆∞·ªõc ·∫£nh: {img_size}")
show_status(f"Batch size: {batch_size}")

#-------------------------------------------------------------------------------
# B∆∞·ªõc 7: T·∫£i m√¥ h√¨nh YOLOv8 pre-trained
#-------------------------------------------------------------------------------
show_status(f"ƒêang t·∫£i m√¥ h√¨nh YOLOv8 pre-trained: {model_base_name}...")
try:
    model = YOLO(model_base_name)
    show_status("ƒê√£ t·∫£i m√¥ h√¨nh pre-trained th√†nh c√¥ng!")
except Exception as e:
    show_status(f"L·ªói khi t·∫£i m√¥ h√¨nh pre-trained: {e}", is_error=True)
    show_status("Ki·ªÉm tra l·∫°i t√™n model size ho·∫∑c k·∫øt n·ªëi m·∫°ng.", is_warning=True)
    exit()

#-------------------------------------------------------------------------------
# B∆∞·ªõc 8: B·∫Øt ƒë·∫ßu hu·∫•n luy·ªán m√¥ h√¨nh VND
#-------------------------------------------------------------------------------
show_status("B·∫Øt ƒë·∫ßu qu√° tr√¨nh hu·∫•n luy·ªán m√¥ h√¨nh nh·∫≠n d·∫°ng ti·ªÅn VND...")
show_status("Qu√° tr√¨nh n√†y c√≥ th·ªÉ m·∫•t th·ªùi gian. B·∫°n c√≥ th·ªÉ theo d√µi ti·∫øn tr√¨nh b√™n d∆∞·ªõi.")

# Ki·ªÉm tra GPU
device = 0 if torch.cuda.is_available() else 'cpu'
show_status(f"S·ª≠ d·ª•ng thi·∫øt b·ªã: {'GPU' if device == 0 else 'CPU'}")
if device == 'cpu':
    show_status("Hu·∫•n luy·ªán tr√™n CPU s·∫Ω r·∫•t ch·∫≠m. N√™n s·ª≠ d·ª•ng GPU (Runtime > Change runtime type > GPU).", is_warning=True)

# ƒê·∫∑t t√™n cho th∆∞ m·ª•c k·∫øt qu·∫£ hu·∫•n luy·ªán
training_run_name = f'vnd_yolov8{model_size}_{epochs}e'

try:
    results = model.train(
        data=path_to_data_yaml,
        epochs=epochs,
        imgsz=img_size,
        batch=batch_size,
        name=training_run_name,
        project='/content/runs/detect',
        plots=True,
        device=device,
        patience=20,
        save_period=10
    )
    show_status("Hu·∫•n luy·ªán ho√†n t·∫•t!")
    actual_results_output_path = results.save_dir
    show_status(f"K·∫øt qu·∫£ th·ª±c t·∫ø ƒë∆∞·ª£c l∆∞u t·∫°i: {actual_results_output_path}")
except Exception as e:
    show_status(f"L·ªói trong qu√° tr√¨nh hu·∫•n luy·ªán: {e}", is_error=True)
    # In th√™m traceback ƒë·ªÉ debug
    import traceback
    traceback.print_exc()
    exit()

#-------------------------------------------------------------------------------
# B∆∞·ªõc 9: ƒê√°nh gi√° m√¥ h√¨nh (Validation)
#-------------------------------------------------------------------------------
# model.val() ƒë√£ ƒë∆∞·ª£c g·ªçi t·ª± ƒë·ªông ·ªü cu·ªëi qu√° tr√¨nh train n·∫øu plots=True v√† c√≥ val set.
# Tuy nhi√™n, b·∫°n c√≥ th·ªÉ g·ªçi l·∫°i t∆∞·ªùng minh n·∫øu mu·ªën.
show_status("ƒê√°nh gi√° m√¥ h√¨nh tr√™n t·∫≠p validation...")
try:
    # ƒê∆∞·ªùng d·∫´n ƒë·∫øn model t·ªët nh·∫•t sau hu·∫•n luy·ªán
    path_to_best_trained_model = os.path.join(actual_results_output_path, 'weights/best.pt')
    if os.path.exists(path_to_best_trained_model):
        model_trained = YOLO(path_to_best_trained_model)
        metrics = model_trained.val()
        show_status(f"mAP50-95: {metrics.box.map:.4f}")
        show_status(f"mAP50: {metrics.box.map50:.4f}")
    else:
        show_status(f"Kh√¥ng t√¨m th·∫•y model 'best.pt' t·∫°i '{path_to_best_trained_model}'. B·ªè qua ƒë√°nh gi√° l·∫°i.", is_warning=True)
except Exception as e:
    show_status(f"L·ªói khi ƒë√°nh gi√° m√¥ h√¨nh: {e}", is_error=True)

#-------------------------------------------------------------------------------
# B∆∞·ªõc 10: Hi·ªÉn th·ªã c√°c bi·ªÉu ƒë·ªì k·∫øt qu·∫£ hu·∫•n luy·ªán
#-------------------------------------------------------------------------------
show_status(f"Hi·ªÉn th·ªã k·∫øt qu·∫£ hu·∫•n luy·ªán t·ª´: {actual_results_output_path}")

# C√°c file h√¨nh ·∫£nh k·∫øt qu·∫£ th∆∞·ªùng ƒë∆∞·ª£c l∆∞u trong th∆∞ m·ª•c n√†y
result_images_to_display = {
    "Bi·ªÉu ƒë·ªì t·ªïng h·ª£p k·∫øt qu·∫£": "results.png",
    "Ma tr·∫≠n nh·∫ßm l·∫´n": "confusion_matrix.png",
    "ƒê∆∞·ªùng cong P": "P_curve.png",
    "ƒê∆∞·ªùng cong R": "R_curve.png",
    "ƒê∆∞·ªùng cong PR": "PR_curve.png",
    "ƒê∆∞·ªùng cong F1": "F1_curve.png"
    # Th√™m c√°c file kh√°c n·∫øu c√≥, v√≠ d·ª•: labels.jpg, labels_correlogram.jpg
}

for title, img_name in result_images_to_display.items():
    img_path = os.path.join(actual_results_output_path, img_name)
    if os.path.exists(img_path):
        show_status(f"\n--- {title} ---")
        try:
            display(Image(filename=img_path, width=700))
        except Exception as e:
            show_status(f"Kh√¥ng th·ªÉ hi·ªÉn th·ªã '{img_name}': {e}", is_warning=True)
    else:
        show_status(f"Kh√¥ng t√¨m th·∫•y file '{img_name}' ƒë·ªÉ hi·ªÉn th·ªã.", is_warning=True)

#-------------------------------------------------------------------------------
# B∆∞·ªõc 11: Xu·∫•t m√¥ h√¨nh (T√πy ch·ªçn)
#-------------------------------------------------------------------------------
# show_status("ƒêang th·ª≠ xu·∫•t m√¥ h√¨nh sang c√°c ƒë·ªãnh d·∫°ng kh√°c (v√≠ d·ª•: ONNX)...")
# try:
#     # S·ª≠ d·ª•ng model_trained (model t·ªët nh·∫•t ƒë√£ load) ƒë·ªÉ export
#     if 'model_trained' in locals() and model_trained is not None:
#         onnx_path = model_trained.export(format='onnx', imgsz=img_size, opset=12) # opset c√≥ th·ªÉ c·∫ßn ƒëi·ªÅu ch·ªânh
#         show_status(f"ƒê√£ xu·∫•t m√¥ h√¨nh sang ONNX t·∫°i: {onnx_path}")
#     else:
#         show_status("Kh√¥ng c√≥ model ƒë√£ hu·∫•n luy·ªán ƒë·ªÉ xu·∫•t. B·ªè qua.", is_warning=True)
# except Exception as e:
#     show_status(f"L·ªói khi xu·∫•t m√¥ h√¨nh sang ONNX: {e}", is_error=True)
#     show_status("Th vi·ªác export c√≥ th·ªÉ kh√¥ng c·∫ßn thi·∫øt cho d·ª± √°n c·ªßa b·∫°n.", is_warning=True)

#-------------------------------------------------------------------------------
# B∆∞·ªõc 12: Ki·ªÉm tra m√¥ h√¨nh tr√™n ·∫£nh th·ª≠ nghi·ªám (n·∫øu c√≥)
#-------------------------------------------------------------------------------
test_images_dir_vnd = os.path.join(dataset_base_dir_vnd, "test", "images") # ƒê∆∞·ªùng d·∫´n chu·∫©n
show_status(f"Ki·ªÉm tra m√¥ h√¨nh tr√™n ·∫£nh th·ª≠ nghi·ªám t·ª´: {test_images_dir_vnd}")

if os.path.exists(test_images_dir_vnd) and os.listdir(test_images_dir_vnd):
    test_image_files = [os.path.join(test_images_dir_vnd, f) for f in os.listdir(test_images_dir_vnd) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]

    if test_image_files:
        num_test_images_to_show = min(len(test_image_files), 5) # Hi·ªÉn th·ªã t·ªëi ƒëa 5 ·∫£nh
        selected_test_images = np.random.choice(test_image_files, num_test_images_to_show, replace=False)

        if 'model_trained' not in locals() or model_trained is None:
             path_to_best_trained_model = os.path.join(actual_results_output_path, 'weights/best.pt')
             if os.path.exists(path_to_best_trained_model):
                model_trained = YOLO(path_to_best_trained_model)
             else:
                show_status("Kh√¥ng t√¨m th·∫•y model 'best.pt' ƒë·ªÉ ch·∫°y th·ª≠ nghi·ªám.", is_error=True)
                model_trained = None # ƒê·∫£m b·∫£o bi·∫øn t·ªìn t·∫°i

        if model_trained:
            plt.figure(figsize=(15, num_test_images_to_show * 5))
            for i, img_path in enumerate(selected_test_images):
                try:
                    predictions = model_trained.predict(img_path, conf=0.35) # Ng∆∞·ª°ng conf c√≥ th·ªÉ ƒëi·ªÅu ch·ªânh

                    # results[0].plot() tr·∫£ v·ªÅ ·∫£nh ƒë√£ v·∫Ω bounding box
                    # c·∫ßn chuy·ªÉn BGR (OpenCV default) sang RGB (matplotlib default)
                    img_with_boxes_bgr = predictions[0].plot()
                    img_with_boxes_rgb = cv2.cvtColor(img_with_boxes_bgr, cv2.COLOR_BGR2RGB)

                    plt.subplot(num_test_images_to_show, 1, i + 1)
                    plt.imshow(img_with_boxes_rgb)
                    plt.title(f"D·ª± ƒëo√°n tr√™n: {os.path.basename(img_path)}")
                    plt.axis('off')
                except Exception as e:
                    show_status(f"L·ªói khi d·ª± ƒëo√°n tr√™n ·∫£nh {os.path.basename(img_path)}: {e}", is_error=True)
            plt.tight_layout()
            plt.show()
            show_status("ƒê√£ hi·ªÉn th·ªã k·∫øt qu·∫£ d·ª± ƒëo√°n tr√™n m·ªôt s·ªë ·∫£nh th·ª≠ nghi·ªám.")
    else:
        show_status("Kh√¥ng t√¨m th·∫•y file ·∫£nh n√†o trong th∆∞ m·ª•c th·ª≠ nghi·ªám.", is_warning=True)
else:
    show_status(f"Th∆∞ m·ª•c ·∫£nh th·ª≠ nghi·ªám '{test_images_dir_vnd}' kh√¥ng t·ªìn t·∫°i ho·∫∑c r·ªóng. B·ªè qua b∆∞·ªõc n√†y.", is_warning=True)


#-------------------------------------------------------------------------------
# B∆∞·ªõc 13: L∆∞u m√¥ h√¨nh t·ªët nh·∫•t (best.pt) v√†o Google Drive
#-------------------------------------------------------------------------------
gdrive_save_dir = '/content/drive/MyDrive/VND_Currency_Models' # Th∆∞ m·ª•c l∆∞u tr√™n Google Drive
model_weights_dir_local = os.path.join(actual_results_output_path, 'weights')
best_model_local_path = os.path.join(model_weights_dir_local, 'best.pt')

show_status(f"ƒêang chu·∫©n b·ªã l∆∞u m√¥ h√¨nh 'best.pt' v√†o Google Drive t·∫°i: {gdrive_save_dir}")

if os.path.exists(best_model_local_path):
    try:
        os.makedirs(gdrive_save_dir, exist_ok=True)
        # ƒê·ªïi t√™n file model ƒë·ªÉ c√≥ th√¥ng tin r√µ r√†ng h∆°n (v√≠ d·ª•: vnd_yolov8n_75e_best.pt)
        gdrive_model_filename = f"vnd_yolov8{model_size}_{epochs}e_best.pt"
        gdrive_full_path = os.path.join(gdrive_save_dir, gdrive_model_filename)

        shutil.copy2(best_model_local_path, gdrive_full_path) # copy2 gi·ªØ metadata
        show_status(f"ƒê√£ l∆∞u m√¥ h√¨nh th√†nh c√¥ng v√†o Google Drive: {gdrive_full_path}")

        # (T√πy ch·ªçn) L∆∞u to√†n b·ªô th∆∞ m·ª•c runs/detect/training_run_name v√†o Drive n·∫øu mu·ªën
        # show_status(f"ƒêang sao ch√©p to√†n b·ªô th∆∞ m·ª•c k·∫øt qu·∫£ hu·∫•n luy·ªán v√†o Google Drive...")
        # shutil.copytree(os.path.join('/content/runs/detect', training_run_name),
        #                 os.path.join(gdrive_save_dir, training_run_name),
        #                 dirs_exist_ok=True)
        # show_status(f"ƒê√£ sao ch√©p th∆∞ m·ª•c hu·∫•n luy·ªán v√†o {os.path.join(gdrive_save_dir, training_run_name)}")

    except Exception as e:
        show_status(f"L·ªói khi l∆∞u m√¥ h√¨nh v√†o Google Drive: {e}", is_error=True)
        show_status("H√£y ƒë·∫£m b·∫£o Google Drive ƒë√£ ƒë∆∞·ª£c k·∫øt n·ªëi v√† c√≥ ƒë·ªß dung l∆∞·ª£ng.", is_warning=True)
else:
    show_status(f"Kh√¥ng t√¨m th·∫•y file m√¥ h√¨nh 'best.pt' t·∫°i '{best_model_local_path}'. Kh√¥ng th·ªÉ l∆∞u v√†o Google Drive.", is_error=True)

#-------------------------------------------------------------------------------
# B∆∞·ªõc 14: T·∫£i m√¥ h√¨nh t·ªët nh·∫•t (best.pt) v·ªÅ m√°y t√≠nh c·ª•c b·ªô
#-------------------------------------------------------------------------------
show_status("ƒêang chu·∫©n b·ªã t·∫£i m√¥ h√¨nh 'best.pt' v·ªÅ m√°y t√≠nh c·ªßa b·∫°n...")
if os.path.exists(best_model_local_path):
    try:
        # ƒê·ªïi t√™n file t·∫£i v·ªÅ ƒë·ªÉ d·ªÖ nh·∫≠n bi·∫øt (v√≠ d·ª•: best_vnd_model.pt)
        download_filename = f"best_vnd_yolov8{model_size}_{epochs}e.pt"
        shutil.copy2(best_model_local_path, download_filename)
        files.download(download_filename)
        show_status(f"ƒê√£ t·∫£i file '{download_filename}' v·ªÅ m√°y t√≠nh th√†nh c√¥ng!")
        show_status("B·∫°n c√≥ th·ªÉ t√¨m th·∫•y file n√†y trong th∆∞ m·ª•c Downloads c·ªßa tr√¨nh duy·ªát.")
    except Exception as e:
        show_status(f"L·ªói khi t·∫£i m√¥ h√¨nh v·ªÅ m√°y t√≠nh: {e}", is_error=True)
else:
    show_status(f"Kh√¥ng t√¨m th·∫•y file m√¥ h√¨nh 'best.pt' ƒë·ªÉ t·∫£i v·ªÅ.", is_error=True)

#-------------------------------------------------------------------------------
# K·∫øt th√∫c
#-------------------------------------------------------------------------------
show_status("Ho√†n t·∫•t qu√° tr√¨nh hu·∫•n luy·ªán, ƒë√°nh gi√° v√† l∆∞u tr·ªØ m√¥ h√¨nh nh·∫≠n d·∫°ng ti·ªÅn VND!")
print("\nL·ªùi khuy√™n:")
print("- ƒê·∫∑t t√™n file model ƒë√£ t·∫£i v·ªÅ (v√≠ d·ª•: 'best_vnd.pt') v√† ƒë·∫∑t v√†o th∆∞ m·ª•c 'models' c·ªßa d·ª± √°n Vision-I.")
print("- C·∫≠p nh·∫≠t file 'vision_i_revised.py' ƒë·ªÉ t·∫£i v√† s·ª≠ d·ª•ng m√¥ h√¨nh m·ªõi n√†y.")
print(f"- M√¥ h√¨nh v√† k·∫øt qu·∫£ hu·∫•n luy·ªán (n·∫øu ƒë√£ ch·ªçn l∆∞u) c√≥ th·ªÉ t√¨m th·∫•y trong Google Drive t·∫°i: {gdrive_save_dir}")
```
