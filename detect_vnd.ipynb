```python
#-------------------------------------------------------------------------------
# Bước 0: Các hàm tiện ích và import ban đầu
#-------------------------------------------------------------------------------
import os
import yaml
import shutil
from IPython.display import clear_output, display # Image sẽ được import sau khi cài ultralytics
from google.colab import drive, files
import numpy as np
import cv2 # Sẽ được cài cùng opencv-python từ ultralytics
import zipfile
import matplotlib.pyplot as plt # Sẽ được cài cùng ultralytics

def show_status(message, is_warning=False, is_error=False):
    prefix = "[INFO]"
    if is_warning:
        prefix = "[CẢNH BÁO]"
    elif is_error:
        prefix = "[LỖI]"
    print(f"{prefix} {message}")

#-------------------------------------------------------------------------------
# Bước 1: Cài đặt các gói cần thiết
#-------------------------------------------------------------------------------
show_status("Đang cài đặt các gói cần thiết (ultralytics, torch, torchvision)...")
%pip install ultralytics torch torchvision --quiet
from IPython.display import Image # Import Image sau khi ultralytics (và pillow) được cài
clear_output(wait=True) # Xóa output của pip install
show_status("Đã cài đặt các gói cần thiết!")

# Import các thư viện đã cài đặt
import torch
from ultralytics import YOLO

#-------------------------------------------------------------------------------
# Bước 2: Kết nối Google Drive
#-------------------------------------------------------------------------------
show_status("Đang kết nối Google Drive...")
try:
    drive.mount('/content/drive', force_remount=True) # Thêm force_remount để đảm bảo kết nối lại nếu cần
    clear_output(wait=True)
    show_status("Google Drive đã được kết nối thành công!")
except Exception as e:
    show_status(f"Không thể kết nối Google Drive: {e}", is_error=True)
    show_status("Vui lòng kiểm tra lại quyền truy cập hoặc thử kết nối lại.", is_warning=True)
    # Có thể dừng ở đây hoặc cho phép tiếp tục mà không có Drive
    # exit()

#-------------------------------------------------------------------------------
# Bước 3: Tạo thư mục cho dự án VND
#-------------------------------------------------------------------------------
# Đổi tên thư mục dự án để phân biệt với dự án USD
project_dir_vnd = '/content/vnd_currency_detection'
dataset_base_dir_vnd = os.path.join(project_dir_vnd, 'dataset') # Thư mục chứa data.yaml và các folder train/valid/test

show_status(f"Tạo thư mục dự án tại {project_dir_vnd}")
os.makedirs(dataset_base_dir_vnd, exist_ok=True) # Tạo cả thư mục dataset


#-------------------------------------------------------------------------------
# Bước 4: Upload Dataset Tiền VND
#-------------------------------------------------------------------------------
show_status("Vui lòng upload dataset tiền VND.")

uploaded = files.upload()

main_zip_file = None
data_yaml_file = None
uploaded_file_names = list(uploaded.keys())

for filename in uploaded_file_names:
    if filename.lower().endswith('.zip'): # Kiểm tra không phân biệt hoa thường
        # Giả định file zip chính chứa toàn bộ dataset nếu có tên chung chung
        # Hoặc người dùng chỉ upload 1 file zip
        if main_zip_file is None : # Ưu tiên file zip đầu tiên làm file zip chính
             main_zip_file = filename
        show_status(f"Đã tìm thấy file ZIP: {filename}")
    elif filename.lower() == 'data.yaml':
        data_yaml_file = filename
        # Di chuyển file YAML vào thư mục dataset_base_dir_vnd nếu nó chưa ở đó
        if os.path.dirname(os.path.join(os.getcwd(), filename)) != dataset_base_dir_vnd:
            shutil.move(filename, os.path.join(dataset_base_dir_vnd, 'data.yaml'))
        show_status(f"Đã tìm thấy và di chuyển file YAML: {filename} vào {dataset_base_dir_vnd}")

# Xử lý file ZIP chính nếu có
if main_zip_file:
    show_status(f"Đang giải nén {main_zip_file} vào {dataset_base_dir_vnd}...")
    try:
        with zipfile.ZipFile(main_zip_file, 'r') as zip_ref:
            zip_ref.extractall(dataset_base_dir_vnd) # Giải nén trực tiếp vào thư mục dataset
        show_status(f"Đã giải nén {main_zip_file} thành công!")

        # Sau khi giải nén, kiểm tra xem data.yaml có nằm trong thư mục con không
        # Ví dụ: dataset.zip giải nén ra dataset/data.yaml (thay vì data.yaml trực tiếp)
        # Tìm file data.yaml trong các thư mục con vừa giải nén
        extracted_yaml_path = None
        for root, dirs, files_in_root in os.walk(dataset_base_dir_vnd):
            if 'data.yaml' in files_in_root:
                extracted_yaml_path = os.path.join(root, 'data.yaml')
                break

        if extracted_yaml_path and extracted_yaml_path != os.path.join(dataset_base_dir_vnd, 'data.yaml'):
            # Di chuyển data.yaml ra thư mục dataset_base_dir_vnd nếu nó nằm sâu hơn
            shutil.move(extracted_yaml_path, os.path.join(dataset_base_dir_vnd, 'data.yaml'))
            show_status(f"Đã di chuyển data.yaml từ {extracted_yaml_path} đến thư mục gốc của dataset.")
            # Xóa thư mục rỗng còn lại nếu có (ví dụ: nếu data.yaml nằm trong một thư mục con duy nhất)
            if os.path.isdir(os.path.dirname(extracted_yaml_path)) and not os.listdir(os.path.dirname(extracted_yaml_path)):
                os.rmdir(os.path.dirname(extracted_yaml_path))

        data_yaml_path_check = os.path.join(dataset_base_dir_vnd, 'data.yaml')
        if not os.path.exists(data_yaml_path_check):
            show_status("Không tìm thấy 'data.yaml' trong file ZIP hoặc ở thư mục dataset gốc sau khi giải nén.", is_warning=True)
            show_status("Vui lòng upload 'data.yaml' riêng.")
            uploaded_yaml = files.upload()
            if 'data.yaml' in uploaded_yaml:
                shutil.move('data.yaml', data_yaml_path_check)
                data_yaml_file = 'data.yaml' # Cập nhật lại tên file yaml
                show_status(f"Đã di chuyển 'data.yaml' vào {dataset_base_dir_vnd}")
            else:
                show_status("Không upload file 'data.yaml'. Quá trình có thể gặp lỗi.", is_error=True)
                # exit()
        else:
            data_yaml_file = 'data.yaml' # Đánh dấu là đã có data.yaml

    except zipfile.BadZipFile:
        show_status(f"File {main_zip_file} không phải là file ZIP hợp lệ hoặc bị lỗi.", is_error=True)
        main_zip_file = None # Đặt lại để có thể xử lý theo cách upload file rời
    except Exception as e:
        show_status(f"Lỗi khi giải nén {main_zip_file}: {e}", is_error=True)
        main_zip_file = None

# Nếu không có file ZIP chính HOẶC data.yaml vẫn chưa có
if not data_yaml_file:
    data_yaml_path_check = os.path.join(dataset_base_dir_vnd, 'data.yaml')
    if not os.path.exists(data_yaml_path_check):
        show_status("Chưa có file 'data.yaml'. Vui lòng upload 'data.yaml'.")
        uploaded_yaml = files.upload()
        if 'data.yaml' in uploaded_yaml:
            shutil.move('data.yaml', data_yaml_path_check)
            data_yaml_file = 'data.yaml'
            show_status(f"Đã di chuyển 'data.yaml' vào {dataset_base_dir_vnd}")
        else:
            show_status("BẮT BUỘC phải có 'data.yaml'. Dừng chương trình.", is_error=True)
            exit()
    else:
         data_yaml_file = 'data.yaml' # Đã có sẵn từ trước


#-------------------------------------------------------------------------------
# Bước 5: Kiểm tra và Cập nhật file data.yaml
#-------------------------------------------------------------------------------
path_to_data_yaml = os.path.join(dataset_base_dir_vnd, 'data.yaml')
show_status(f"Kiểm tra file data.yaml tại: {path_to_data_yaml}")

if os.path.exists(path_to_data_yaml):
    show_status("Đọc thông tin từ file data.yaml...")
    try:
        with open(path_to_data_yaml, 'r') as file:
            yaml_content = yaml.safe_load(file)

        if yaml_content is None:
            show_status("File data.yaml rỗng hoặc không đúng định dạng.", is_error=True)
            exit()

        print("\nThông tin classes từ data.yaml:")
        names = yaml_content.get('names', [])
        if names:
            for idx, name in enumerate(names):
                print(f"  {idx}: {name}")
        else:
            show_status("Không tìm thấy danh sách 'names' trong data.yaml.", is_warning=True)

        nc = yaml_content.get('nc')
        if nc is not None:
            print(f"Số lượng classes (nc): {nc}")
            if names and len(names) != nc:
                show_status(f"Số lượng 'names' ({len(names)}) không khớp với 'nc' ({nc}).", is_warning=True)
        else:
            show_status("Không tìm thấy 'nc' (số lượng classes) trong data.yaml.", is_warning=True)


        requires_update = False
        for key in ['train', 'val', 'test']:
            if key in yaml_content:
                current_path = yaml_content[key]
                # Nếu đường dẫn là tương đối và không bắt đầu bằng '../' (chỉ tên thư mục)
                # thì nó được hiểu là nằm trong cùng thư mục với data.yaml
                # Ví dụ: train: train/images

                # Đường dẫn tuyệt đối đến thư mục images/labels mong muốn
                # Ví dụ: /content/vnd_currency_detection/dataset/train/images
                expected_abs_path_dir = os.path.join(dataset_base_dir_vnd, os.path.normpath(current_path))

                if not os.path.exists(expected_abs_path_dir):
                    show_status(f"Đường dẫn '{current_path}' cho '{key}' trong data.yaml không tồn tại tại '{expected_abs_path_dir}'.", is_warning=True)
                    # Có thể thử một vài cấu trúc phổ biến
                    # Ví dụ, nếu data.yaml nằm trong dataset/ và dataset.yaml ghi train: images/ (thiếu train/)
                    common_paths_to_try = [
                        os.path.join(dataset_base_dir_vnd, key, 'images'), # dataset/train/images
                        os.path.join(dataset_base_dir_vnd, os.path.basename(os.path.normpath(current_path))), # dataset/images (nếu current_path là train/images)
                    ]
                    found_alternative = False
                    for alt_path_dir in common_paths_to_try:
                        if os.path.exists(alt_path_dir) and os.path.isdir(alt_path_dir):
                            # Cập nhật lại đường dẫn tương đối cho data.yaml
                            # ví dụ: từ dataset_base_dir_vnd, đường dẫn đến alt_path_dir là gì?
                            relative_new_path = os.path.relpath(alt_path_dir, start=dataset_base_dir_vnd)
                            yaml_content[key] = relative_new_path # Sửa thành đường dẫn tương đối đúng
                            show_status(f"Đã cập nhật đường dẫn cho '{key}' thành '{relative_new_path}' (trỏ đến {alt_path_dir})")
                            requires_update = True
                            found_alternative = True
                            break
                    if not found_alternative:
                         show_status(f"Không thể tự động sửa đường dẫn cho '{key}'. Vui lòng kiểm tra thủ công file data.yaml.", is_error=True)

        if requires_update:
            with open(path_to_data_yaml, 'w') as outfile:
                yaml.dump(yaml_content, outfile, default_flow_style=False, sort_keys=False)
            show_status("File data.yaml đã được cập nhật với đường dẫn mới (nếu có).")

        # In lại cấu trúc thư mục dataset để kiểm tra
        show_status("Cấu trúc thư mục dataset hiện tại:")
        # !ls -R {dataset_base_dir_vnd} # ls -R có thể quá dài, chỉ hiện cấp đầu
        for item in os.listdir(dataset_base_dir_vnd):
            print(f"  └── {item}")
            if os.path.isdir(os.path.join(dataset_base_dir_vnd, item)):
                 for sub_item in os.listdir(os.path.join(dataset_base_dir_vnd, item)):
                      print(f"      └── {sub_item}")


    except yaml.YAMLError as e:
        show_status(f"Lỗi khi đọc file data.yaml: {e}", is_error=True)
        exit()
    except Exception as e:
        show_status(f"Lỗi không xác định khi xử lý data.yaml: {e}", is_error=True)
        exit()
else:
    show_status(f"KHÔNG TÌM THẤY file data.yaml tại '{path_to_data_yaml}'!", is_error=True)
    show_status("Vui lòng upload file data.yaml vào thư mục dataset và chạy lại.", is_warning=True)
    exit()

#-------------------------------------------------------------------------------
# Bước 6: Thiết lập các tham số huấn luyện
#-------------------------------------------------------------------------------
show_status("Thiết lập các tham số huấn luyện cho mô hình VND...")

# Giá trị mặc định
default_epochs = 75 # Giảm một chút so với 100 nếu dataset nhỏ hoặc để test nhanh
default_img_size = 640
default_batch_size = 16 # Phổ biến cho Colab free tier
default_model_size = 'n' # nano (nhẹ nhất), s (small), m (medium)

print("\nBạn có thể tùy chỉnh các tham số huấn luyện (nhấn Enter để dùng giá trị mặc định):")
try:
    epochs_input = input(f"1. Số lượng epochs (mặc định: {default_epochs}): ")
    epochs = int(epochs_input) if epochs_input else default_epochs

    img_size_input = input(f"2. Kích thước ảnh (mặc định: {default_img_size}): ")
    img_size = int(img_size_input) if img_size_input else default_img_size

    batch_size_input = input(f"3. Kích thước batch (mặc định: {default_batch_size}): ")
    batch_size = int(batch_size_input) if batch_size_input else default_batch_size

    model_size_input = input(f"4. Model size (n, s, m, l, x - mặc định: {default_model_size}): ").lower()
    model_size = model_size_input if model_size_input in ['n', 's', 'm', 'l', 'x'] else default_model_size
except ValueError:
    show_status("Đầu vào không hợp lệ, sử dụng giá trị mặc định.", is_warning=True)
    epochs, img_size, batch_size, model_size = default_epochs, default_img_size, default_batch_size, default_model_size


model_base_name = f'yolov8{model_size}.pt'
show_status(f"Sử dụng mô hình cơ sở: {model_base_name}")
show_status(f"Số lượng epochs: {epochs}")
show_status(f"Kích thước ảnh: {img_size}")
show_status(f"Batch size: {batch_size}")

#-------------------------------------------------------------------------------
# Bước 7: Tải mô hình YOLOv8 pre-trained
#-------------------------------------------------------------------------------
show_status(f"Đang tải mô hình YOLOv8 pre-trained: {model_base_name}...")
try:
    model = YOLO(model_base_name)
    show_status("Đã tải mô hình pre-trained thành công!")
except Exception as e:
    show_status(f"Lỗi khi tải mô hình pre-trained: {e}", is_error=True)
    show_status("Kiểm tra lại tên model size hoặc kết nối mạng.", is_warning=True)
    exit()

#-------------------------------------------------------------------------------
# Bước 8: Bắt đầu huấn luyện mô hình VND
#-------------------------------------------------------------------------------
show_status("Bắt đầu quá trình huấn luyện mô hình nhận dạng tiền VND...")
show_status("Quá trình này có thể mất thời gian. Bạn có thể theo dõi tiến trình bên dưới.")

# Kiểm tra GPU
device = 0 if torch.cuda.is_available() else 'cpu'
show_status(f"Sử dụng thiết bị: {'GPU' if device == 0 else 'CPU'}")
if device == 'cpu':
    show_status("Huấn luyện trên CPU sẽ rất chậm. Nên sử dụng GPU (Runtime > Change runtime type > GPU).", is_warning=True)

# Đặt tên cho thư mục kết quả huấn luyện
training_run_name = f'vnd_yolov8{model_size}_{epochs}e'

try:
    results = model.train(
        data=path_to_data_yaml,
        epochs=epochs,
        imgsz=img_size,
        batch=batch_size,
        name=training_run_name,
        project='/content/runs/detect',
        plots=True,
        device=device,
        patience=20,
        save_period=10
    )
    show_status("Huấn luyện hoàn tất!")
    actual_results_output_path = results.save_dir
    show_status(f"Kết quả thực tế được lưu tại: {actual_results_output_path}")
except Exception as e:
    show_status(f"Lỗi trong quá trình huấn luyện: {e}", is_error=True)
    # In thêm traceback để debug
    import traceback
    traceback.print_exc()
    exit()

#-------------------------------------------------------------------------------
# Bước 9: Đánh giá mô hình (Validation)
#-------------------------------------------------------------------------------
# model.val() đã được gọi tự động ở cuối quá trình train nếu plots=True và có val set.
# Tuy nhiên, bạn có thể gọi lại tường minh nếu muốn.
show_status("Đánh giá mô hình trên tập validation...")
try:
    # Đường dẫn đến model tốt nhất sau huấn luyện
    path_to_best_trained_model = os.path.join(actual_results_output_path, 'weights/best.pt')
    if os.path.exists(path_to_best_trained_model):
        model_trained = YOLO(path_to_best_trained_model)
        metrics = model_trained.val()
        show_status(f"mAP50-95: {metrics.box.map:.4f}")
        show_status(f"mAP50: {metrics.box.map50:.4f}")
    else:
        show_status(f"Không tìm thấy model 'best.pt' tại '{path_to_best_trained_model}'. Bỏ qua đánh giá lại.", is_warning=True)
except Exception as e:
    show_status(f"Lỗi khi đánh giá mô hình: {e}", is_error=True)

#-------------------------------------------------------------------------------
# Bước 10: Hiển thị các biểu đồ kết quả huấn luyện
#-------------------------------------------------------------------------------
show_status(f"Hiển thị kết quả huấn luyện từ: {actual_results_output_path}")

# Các file hình ảnh kết quả thường được lưu trong thư mục này
result_images_to_display = {
    "Biểu đồ tổng hợp kết quả": "results.png",
    "Ma trận nhầm lẫn": "confusion_matrix.png",
    "Đường cong P": "P_curve.png",
    "Đường cong R": "R_curve.png",
    "Đường cong PR": "PR_curve.png",
    "Đường cong F1": "F1_curve.png"
    # Thêm các file khác nếu có, ví dụ: labels.jpg, labels_correlogram.jpg
}

for title, img_name in result_images_to_display.items():
    img_path = os.path.join(actual_results_output_path, img_name)
    if os.path.exists(img_path):
        show_status(f"\n--- {title} ---")
        try:
            display(Image(filename=img_path, width=700))
        except Exception as e:
            show_status(f"Không thể hiển thị '{img_name}': {e}", is_warning=True)
    else:
        show_status(f"Không tìm thấy file '{img_name}' để hiển thị.", is_warning=True)

#-------------------------------------------------------------------------------
# Bước 11: Xuất mô hình (Tùy chọn)
#-------------------------------------------------------------------------------
# show_status("Đang thử xuất mô hình sang các định dạng khác (ví dụ: ONNX)...")
# try:
#     # Sử dụng model_trained (model tốt nhất đã load) để export
#     if 'model_trained' in locals() and model_trained is not None:
#         onnx_path = model_trained.export(format='onnx', imgsz=img_size, opset=12) # opset có thể cần điều chỉnh
#         show_status(f"Đã xuất mô hình sang ONNX tại: {onnx_path}")
#     else:
#         show_status("Không có model đã huấn luyện để xuất. Bỏ qua.", is_warning=True)
# except Exception as e:
#     show_status(f"Lỗi khi xuất mô hình sang ONNX: {e}", is_error=True)
#     show_status("Th việc export có thể không cần thiết cho dự án của bạn.", is_warning=True)

#-------------------------------------------------------------------------------
# Bước 12: Kiểm tra mô hình trên ảnh thử nghiệm (nếu có)
#-------------------------------------------------------------------------------
test_images_dir_vnd = os.path.join(dataset_base_dir_vnd, "test", "images") # Đường dẫn chuẩn
show_status(f"Kiểm tra mô hình trên ảnh thử nghiệm từ: {test_images_dir_vnd}")

if os.path.exists(test_images_dir_vnd) and os.listdir(test_images_dir_vnd):
    test_image_files = [os.path.join(test_images_dir_vnd, f) for f in os.listdir(test_images_dir_vnd) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]

    if test_image_files:
        num_test_images_to_show = min(len(test_image_files), 5) # Hiển thị tối đa 5 ảnh
        selected_test_images = np.random.choice(test_image_files, num_test_images_to_show, replace=False)

        if 'model_trained' not in locals() or model_trained is None:
             path_to_best_trained_model = os.path.join(actual_results_output_path, 'weights/best.pt')
             if os.path.exists(path_to_best_trained_model):
                model_trained = YOLO(path_to_best_trained_model)
             else:
                show_status("Không tìm thấy model 'best.pt' để chạy thử nghiệm.", is_error=True)
                model_trained = None # Đảm bảo biến tồn tại

        if model_trained:
            plt.figure(figsize=(15, num_test_images_to_show * 5))
            for i, img_path in enumerate(selected_test_images):
                try:
                    predictions = model_trained.predict(img_path, conf=0.35) # Ngưỡng conf có thể điều chỉnh

                    # results[0].plot() trả về ảnh đã vẽ bounding box
                    # cần chuyển BGR (OpenCV default) sang RGB (matplotlib default)
                    img_with_boxes_bgr = predictions[0].plot()
                    img_with_boxes_rgb = cv2.cvtColor(img_with_boxes_bgr, cv2.COLOR_BGR2RGB)

                    plt.subplot(num_test_images_to_show, 1, i + 1)
                    plt.imshow(img_with_boxes_rgb)
                    plt.title(f"Dự đoán trên: {os.path.basename(img_path)}")
                    plt.axis('off')
                except Exception as e:
                    show_status(f"Lỗi khi dự đoán trên ảnh {os.path.basename(img_path)}: {e}", is_error=True)
            plt.tight_layout()
            plt.show()
            show_status("Đã hiển thị kết quả dự đoán trên một số ảnh thử nghiệm.")
    else:
        show_status("Không tìm thấy file ảnh nào trong thư mục thử nghiệm.", is_warning=True)
else:
    show_status(f"Thư mục ảnh thử nghiệm '{test_images_dir_vnd}' không tồn tại hoặc rỗng. Bỏ qua bước này.", is_warning=True)


#-------------------------------------------------------------------------------
# Bước 13: Lưu mô hình tốt nhất (best.pt) vào Google Drive
#-------------------------------------------------------------------------------
gdrive_save_dir = '/content/drive/MyDrive/VND_Currency_Models' # Thư mục lưu trên Google Drive
model_weights_dir_local = os.path.join(actual_results_output_path, 'weights')
best_model_local_path = os.path.join(model_weights_dir_local, 'best.pt')

show_status(f"Đang chuẩn bị lưu mô hình 'best.pt' vào Google Drive tại: {gdrive_save_dir}")

if os.path.exists(best_model_local_path):
    try:
        os.makedirs(gdrive_save_dir, exist_ok=True)
        # Đổi tên file model để có thông tin rõ ràng hơn (ví dụ: vnd_yolov8n_75e_best.pt)
        gdrive_model_filename = f"vnd_yolov8{model_size}_{epochs}e_best.pt"
        gdrive_full_path = os.path.join(gdrive_save_dir, gdrive_model_filename)

        shutil.copy2(best_model_local_path, gdrive_full_path) # copy2 giữ metadata
        show_status(f"Đã lưu mô hình thành công vào Google Drive: {gdrive_full_path}")

        # (Tùy chọn) Lưu toàn bộ thư mục runs/detect/training_run_name vào Drive nếu muốn
        # show_status(f"Đang sao chép toàn bộ thư mục kết quả huấn luyện vào Google Drive...")
        # shutil.copytree(os.path.join('/content/runs/detect', training_run_name),
        #                 os.path.join(gdrive_save_dir, training_run_name),
        #                 dirs_exist_ok=True)
        # show_status(f"Đã sao chép thư mục huấn luyện vào {os.path.join(gdrive_save_dir, training_run_name)}")

    except Exception as e:
        show_status(f"Lỗi khi lưu mô hình vào Google Drive: {e}", is_error=True)
        show_status("Hãy đảm bảo Google Drive đã được kết nối và có đủ dung lượng.", is_warning=True)
else:
    show_status(f"Không tìm thấy file mô hình 'best.pt' tại '{best_model_local_path}'. Không thể lưu vào Google Drive.", is_error=True)

#-------------------------------------------------------------------------------
# Bước 14: Tải mô hình tốt nhất (best.pt) về máy tính cục bộ
#-------------------------------------------------------------------------------
show_status("Đang chuẩn bị tải mô hình 'best.pt' về máy tính của bạn...")
if os.path.exists(best_model_local_path):
    try:
        # Đổi tên file tải về để dễ nhận biết (ví dụ: best_vnd_model.pt)
        download_filename = f"best_vnd_yolov8{model_size}_{epochs}e.pt"
        shutil.copy2(best_model_local_path, download_filename)
        files.download(download_filename)
        show_status(f"Đã tải file '{download_filename}' về máy tính thành công!")
        show_status("Bạn có thể tìm thấy file này trong thư mục Downloads của trình duyệt.")
    except Exception as e:
        show_status(f"Lỗi khi tải mô hình về máy tính: {e}", is_error=True)
else:
    show_status(f"Không tìm thấy file mô hình 'best.pt' để tải về.", is_error=True)

#-------------------------------------------------------------------------------
# Kết thúc
#-------------------------------------------------------------------------------
show_status("Hoàn tất quá trình huấn luyện, đánh giá và lưu trữ mô hình nhận dạng tiền VND!")
print("\nLời khuyên:")
print("- Đặt tên file model đã tải về (ví dụ: 'best_vnd.pt') và đặt vào thư mục 'models' của dự án Vision-I.")
print("- Cập nhật file 'vision_i_revised.py' để tải và sử dụng mô hình mới này.")
print(f"- Mô hình và kết quả huấn luyện (nếu đã chọn lưu) có thể tìm thấy trong Google Drive tại: {gdrive_save_dir}")
```
